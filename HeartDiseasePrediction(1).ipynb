```python
df_cleaned.head()
```

```python
from scipy.stats import chi2_contingency
alpha=0.05
result={}
for col in non_numeric_columns:
  conti_tab=pd.crosstab(df_cleaned[col],df_cleaned["HeartDisease"])
  chi_stats,p_val,_,_=chi2_contingency(conti_tab)
  decision="keep feature" if p_val<alpha else "remove feature"
  result[col]={
      "chi_stats":chi_stats,
      "p_val":p_val,
      "decision":decision
  }
print(result)

```

```python
df_enc=pd.read_csv("heart.xls")
```

```python
df_enc=pd.get_dummies(df_enc)
```

```python
df_enc=df_enc.astype(int)
```

```python
ch_mean=df_enc.loc[df_enc['Cholesterol'] !=0,'Cholesterol'].mean()
df_enc["Cholesterol"]=df_enc["Cholesterol"].replace({0:ch_mean})
ch_mean=df_enc.loc[df_enc['FastingBS'] !=0,'FastingBS'].mean()
df_enc["FastingBS"]=df_enc["FastingBS"].replace({0:ch_mean})
```

```python
df_enc
```

```python
chi_columns=['Sex_F', 'Sex_M', 'ChestPainType_ASY',
       'ChestPainType_ATA', 'ChestPainType_NAP', 'ChestPainType_TA',
       'RestingECG_LVH', 'RestingECG_Normal', 'RestingECG_ST',
       'ExerciseAngina_N', 'ExerciseAngina_Y', 'ST_Slope_Down',
       'ST_Slope_Flat', 'ST_Slope_Up']
```

```python
from scipy.stats import chi2_contingency
alpha=0.05
result={}
for col in chi_columns:
  conti_tab=pd.crosstab(df_enc[col],df_enc["HeartDisease"])
  chi_stats,p_val,_,_=chi2_contingency(conti_tab)
  decision="keep feature" if p_val<alpha else "remove feature"
  result[col]={
      "chi_stats":chi_stats,
      "p_val":p_val,
      "decision":decision
  }
print(result)

```

```python
result_table_enc=pd.DataFrame(result).T
```
